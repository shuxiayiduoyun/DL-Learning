```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, KBinsDiscretizer, LabelEncoder, OrdinalEncoder
```

```python
data = {
    'age': [10, 15, 12, 45, 36],
    'gender': ['男', '女', '男', '男', '女'],
    'major': ['计算机', '软件工程', '物理', '计算机', '数学']}
data = pd.DataFrame(data)
data
```

```python
ct = ColumnTransformer([
    ('cate_feat', OneHotEncoder(sparse_output=False), ['gender', 'major'])
])
ct.fit_transform(data)
data_trans = ct.transform(data)
data_trans
```
